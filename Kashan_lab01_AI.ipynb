```python
def check_odd_even(number):
  if number % 2 == 0:
    return "Even"
  else:
    return "Odd"

num = int(input("Enter a number:"))
result = check_odd_even(num)
print(f"The number {num} is {result}.")
```

```python
def is_prime(num):
    if num < 2:
        return False
    for j in range(2, int(num**0.5) + 1):
        if num % j == 0:
            return False
    return True

count = 0
for i in range(100):
    if is_prime(i):
        count += 1

print("Total prime numbers up to 99:", count)
```

```python
def is_prime(number):
  if number <= 1:
    return False
  for i in range(2, int(number**0.5) + 1):
    if number % i == 0:
      return False
  return True

num = int(input("Enter a number:"))
if is_prime(num):
  print(f"The number {num} is prime.")
else:
  print(f"The number {num} is not prime.")
```
