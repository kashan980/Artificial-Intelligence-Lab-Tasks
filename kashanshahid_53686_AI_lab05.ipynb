Lab Task 1: Depth First Search (DFS)

```python
def dfs(node, graph, visited, component):
    visited[node] = True
    component.append(node)

    print(f"Visited Node: {node} | Adjacent Nodes: {graph[node]}")

    for child in graph[node]:
        if not visited[child]:
            dfs(child, graph, visited, component)

if __name__ == "__main__":
    graph = {
        0: [1, 2, 3],
        1: [0],
        2: [0, 3, 4],
        3: [0, 2],
        4: [2]
    }

    node = 0
    visited = [False] * len(graph)
    component = []

    print("--- DFS Node Traversal & Adjacency Info ---")
    dfs(node, graph, visited, component)

    print("\nFinal DFS Traversal Order:", component)
```

Lab Task 2: Breadth-First Search (BFS)

```python
from collections import deque

def bfs(start_node, graph):
    visited = [False] * len(graph)
    queue = deque([start_node])
    visited[start_node] = True
    traversal_order = []

    print("--- BFS Node Traversal & Adjacency Info ---")

    while queue:
        node = queue.popleft()
        traversal_order.append(node)

        print(f"Visited Node: {node} | Adjacent Nodes: {graph[node]}")

        for child in graph[node]:
            if not visited[child]:
                visited[child] = True
                queue.append(child)

    return traversal_order

if __name__ == "__main__":
    graph = {
        0: [1, 2, 3],
        1: [0],
        2: [0, 3, 4],
        3: [0, 2],
        4: [2]
    }

    component = bfs(0, graph)
    print("\nFinal BFS Traversal Order:", component)
```
