12. Bonus Task GPA Calculator

```python

num_courses = int(input("Enter the number of courses: "))

total_points = 0
total_credits = 0

course_credits = []
course_marks = []

for i in range(num_courses):
    print(f"\nDetails for Course {i+1}:")

    credit = int(input(f"  Enter credit hours: "))
    marks = float(input(f"  Enter obtained marks: "))

    course_credits.append(credit)
    course_marks.append(marks)

    if marks >= 80:
        gp = 4.0
    elif marks >= 70:
        gp = 3.0
    elif marks >= 60:
        gp = 2.0
    elif marks >= 50:
        gp = 1.0
    else:
        gp = 0.0

    total_points += (gp * credit)
    total_credits += credit


if total_credits > 0:
    semester_gpa = total_points / total_credits
    print("\n------------------------------")
    print(f"Total Credit Hours: {total_credits}")
    print(f"Your Semester GPA is: {semester_gpa:.2f}")
    print("------------------------------")
else:
    print("No credits entered.")
```
